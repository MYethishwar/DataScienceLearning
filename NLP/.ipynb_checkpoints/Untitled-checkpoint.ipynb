```python
import re
text = "Hi hello 0923692 $%367 how are YOU"
```

```python
text = text.lower()
```

```python
text = re.sub(r"[^\w\s]","",text)
print(text)
```

```python
# remove numbers

text = re.sub(r"\d+","",text)
print(text)
```

```python
# Tokenize (split)
tokens = text.split()
print(tokens)
```

```python
# Using NLTK (natural lang tool kit)

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
text = "Hu how are you 9873429ty %$#@%& hi HELLO YETHISHWar"
```

```python
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word not in stop_words and word.isalpha()]

print(filtered_words)
```

```python
# lemmatizer

lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in filtered_words]
lemmas
```

```python
# Stemmming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

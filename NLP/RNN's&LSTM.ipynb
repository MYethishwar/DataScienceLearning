```python
sentences = [
    'I love this product',
    'This is a bad product',
    'Amazing experience',
    'Worst Experience'
]

labels = [1,0,1,0]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences,padding='post')

print(padded)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
x = padded
y = np.array(labels)
```

```python
model = Sequential()
model.add(Embedding(input_dim=50,output_dim=16))
model.add(SimpleRNN(32))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x,y,epochs=30)
```

```python
# SimpleRNN has a problem "Vanishing Gradients"
```

```python
from tensorflow.keras.layers import LSTM

model = Sequential()
model.add(Embedding(input_dim=1000,output_dim=64))
model.add(LSTM(64))
model.add(Dense(1,activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x,y,epochs=30)
```

```python
import pandas as pd

df = pd.read_csv('Resume.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df['Category_encoding'] = encoder.fit_transform(df['Category'])
```

```python
df.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words=5000,oov_token="<OOV>")
tokenizer.fit_on_texts(df['Resume'])
```

```python
sequences = tokenizer.texts_to_sequences(df['Resume'])
padded = pad_sequences(sequences,maxlen=300,padding='post')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(padded,df['Category_encoding'],test_size=0.2,random_state=56)
```

```python
# building an lstm model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,Dropout
```

```python
model = Sequential()
model.add(Embedding(input_dim=5000,output_dim=128))
model.add(LSTM(128,return_sequences=True))
model.add(LSTM(64))
model.add(Dropout(0.5))
model.add(Dense(len(df['Category'].unique()),activation='softmax'))
```

```python
model.compile(
    optimizer='adam',
    metrics=['accuracy'],
    loss='sparse_categorical_crossentropy'
)
```

```python
model.summary()
```

```python
model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=(x_test,y_test)
)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('Iris')
```

```python
df.head()
```

```python
x = df.drop(['species'], axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 97)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=5)

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
# To freeze thw file 
import pickle

pickle.dump(model,open("DecisonTreemodelIris.pkl","wb"))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(40,10))
plot_tree(model, filled=True, class_names=model.classes_)
plt.show()
```

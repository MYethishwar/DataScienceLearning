```python
median_B = data['BloodPressure'].median()
data['BloodPressure'] = data['BloodPressure'].replace(0, median_B)
```

```python
median_I = data['Insulin'].median()
data['Insulin'] = data['Insulin'].replace(0, median_I)
```

```python
median_BMI = data['BMI'].median()
data['BMI'] = data['BMI'].replace(0, median_BMI)
```

```python
x = data.drop('Outcome', axis = 1)
y = data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, random_state=100, test_size = 0.2)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
y_test.value_counts().sum()
```

## Decision Tree Classifier

```python
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier()

modelDT.fit(x_train,y_train)
y_pred = modelDT.predict(x_test)
print(classification_report(y_pred,y_test))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators=5)

modelRF.fit(x_train,y_train)
y_pred = modelRF.predict(x_test)
print(classification_report(y_pred,y_test))
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression()

modelLR.fit(x_train,y_train)
y_pred = modelLR.predict(x_test)
print(classification_report(y_pred,y_test))
```

## Naive Bayes 

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
ypred = model.predict(x_test)
print(classification_report(ypred,y_test))
```

## K - Nearest Neigbours

```python
from sklearn.neighbors import KNeighborsClassifier
modelKnn = KNeighborsClassifier(n_neighbors=5)
modelKnn.fit(x_train,y_train)
ypred = modelKnn.predict(x_test)
print(classification_report(ypred,y_test))
```

## Support Vector Machines

```python
from sklearn.svm import SVC
modelSVC = SVC()
modelSVC.fit(x_train,y_train)
ypred = modelSVC.predict(x_test)
print(classification_report(ypred,y_test))
```

## Hyper Parameter Tuning

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ("scaler",StandardScaler()),
    ("svm",SVC())
])
```

```python
param_grid = {
    "svm__kernel":["linear","rbf","sigmoid"],
    "svm__C":[0.1, 1.0, 10.0, 100.0],
    "svm__gamma":['scale','auto']
}
```

```python
grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv = 5
)
```

```python
grid.fit(x_train,y_train)
```

```python
print("Best Parameters: ",grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
print(classification_report(y_test,y_pred))
```

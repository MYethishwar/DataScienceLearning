```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
df = {
    'BedRooms':[1,2,3,4,5],
    'RentPrice':[10000,12000,15000,30000, 40000]
}
df = pd.DataFrame(df)
```

```python
df.head()
```

```python
import seaborn as sns
sns.scatterplot(data=df, x = 'BedRooms', y = 'RentPrice')
```

```python
df.corr()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = df[['BedRooms']]
y = df['RentPrice']

model.fit(x,y)
```

```python
model.predict([[6]])
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
def prediction(x):
  return (m*x)+c

prediction(9)
```

```python
# Multiple Linear Regression
df = {
    'Bedrooms':[1,2,3,4,5],
    'Bathrooms':[1,1,2,2,3],
    'Area':[800,1000,1400,1600,1700],
    'RentPrice':[10000,12000,15000,20000,25000]
}

df = pd.DataFrame(df)
df.head()
```

```python
x = df[['Bedrooms','Bathrooms','Area']]
y = df['RentPrice']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x,y)
```

```python
model.predict([[1,1,1000]])
```

```python
m1,m2,m3 = model.coef_
c = model.intercept_
```

```python
def predictMLR(x1,x2,x3):
  return ((m1*x1)+(m2*x2)+(m3*x3)+c)

predictMLR(1,1,1500)
```

```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
x = df.drop(['species'],axis=1)
y = df['species']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=71,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ypred,y_test))
```

```python
# K- Nearest Neighbours
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelKnn = KNeighborsClassifier(n_neighbors=5,n_jobs=-1)
```

```python
modelKnn.fit(x_train,y_train)
```

```python
ypred = modelKnn.predict(x_test)
```

```python
print(classification_report(ypred,y_test))
```

```python
# Applying these algorithms with the titanic dataset
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# lets consider these as inputs 
# ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked', 'adult_male', 'alone']
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df['adult_male'] = df['adult_male'].astype(int)
df['alone'] = df['alone'].astype(int)
```

```python
df = pd.get_dummies(df, columns=['sex','embarked'], drop_first=True)
```

```python
x = df[['pclass','age','sibsp','parch','fare','adult_male','alone','sex_male','embarked_Q','embarked_S']]
y = df['alive']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=18)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
import keras
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
plt.matshow(x_train[0])
```

```python
x_train[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
```

```python
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation='relu'),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(
    optimizer ='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=15,validation_data=(x_test,y_test))
```

```python
plt.matshow(x_train[101])
```

```python
import numpy as np
predicted_probabilities = model.predict(np.expand_dims(x_train[101], axis=0))
print(predicted_probabilities)
```

```python
predicted_digit = np.argmax(predicted_probabilities[0])
print(f"The predicted digit for x_train[101] is: {predicted_digit}")
```

```python
sum = 0
for i in predicted_probabilities:
  for j in i:
    sum = sum + j
print(sum)
```

```python
def prediction_img(img):
  predicted_probabilities = model.predict(np.expand_dims(img, axis=0))
  predicted_digit = np.argmax(predicted_probabilities[0])
  print(f"The predicted digit is: {predicted_digit}")
  print("Original Image: ")
  plt.matshow(img)

prediction_img(x_train[9897])
```

```python
# Cleaning Data
# Bad data could be:
#       Empty cells
#       Data in wrong format
#       Wrong data
#       Duplicates
```

```python
# 1. CLeaning Empty cells
```

```python
# One way deal with empty cells is to remove rows that contain empty cells
df = pd.read_csv('dirtydata.csv')
```

```python
print(df.to_string())
```

```python
new_df = df.dropna()
```

```python
# By default the dropna() will not change the orignal dataset instead it will create another one
# so using inplace = True will change the original dataset
df.dropna(inplace = True)
```

```python
# Another way to deal with empty values is to replace with a new appropiate value
df = pd.read_csv('dirtydata.csv') #Again raw data is imported
# fillna() is used to replace empty cells with these values
```

```python
new_df = df.fillna(1000)  #inplace = true to change the original dataset
```

```python
# used to count the null values in the dataframe
df.isnull().sum()
```

```python
#Replace only for specified columns
df.fillna({'Calories':130})
df.fillna({'Date':'2020/12/22'})
```

```python
# Replacing with the mean() median() and mode()
df.fillna({'Calories':df['Calories'].mean()}, inplace= True)
```

```python
# mean() is used when the data is normally distributed
# median() is used when the data is skewed or have outliers
```

```python
# Cleaning data of wrong format
# it is difficult to analyze the data which is in wrong format
# TO fix this we have 2 options
#         Remove the rows
#         Convert all cells in the columns in the same format


# pandas has a to_datetime() for this
df['Date'] = pd.to_datetime(df['Date'],format='mixed')
```

```python
# Drop the rows in the Date column which has null values
df.dropna(subset='Date').head()
```

```python
mydata = pd.DataFrame({'name': ['A','B','C','D','E','F'],
                      'age': [190,25,35, 250, 8,150]})
```

```python
# Identify the outliers using boxplot and z score
import seaborn as sns
sns.boxplot(mydata['age'])
```

```python
# We can also find out the outliers using zscore 
# from scipy import stats
```

```python
# If it is small dataset we can re assign with the value manually
# If we have large dataset then we need to iterate through each value and reassign it 
for i in mydata.index:
    if mydata.loc[i,'age'] > 120:
        mydata.loc[i,'age'] = 100
print(mydata)
```

```python
myarray = np.array([1,2,3,4,5,0])
print(myarray)
new_array = myarray.astype(bool)
print(new_array)
```

```python
copied_array = myarray.copy()
print(f"Original array: {myarray}")
print("")
print(f"Copied array: {copied_array}")
copied_array[0] = 1000
print(f"Edited Copied Array: {copied_array}")
print(f"Original array remains same : {myarray}")
print("")
view_array = myarray.view()
print(f"View of array: {view_array}")
view_array[0] = 1000
print(f"Edited view Array: {view_array}")
print(f"Original array : {myarray}")
```

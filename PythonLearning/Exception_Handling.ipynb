```python
def attempt_float(x):
    try:
        return float(x)
    except:
        return x
attempt_float("yethishwar")
```

```python
def handling_zero_division_error():
    try:
        a = int(input('Enter numerator: '))
        return   2/a
    except ZeroDivisionError:
        return "Entered zero!!!!"
handling_zero_division_error()
```

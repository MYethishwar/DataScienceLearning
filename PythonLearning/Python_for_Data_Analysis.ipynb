```python
# list.append() ->  adds the element at the last
# list.extend() -> appends multiple elements
# list.insert(1) -> inserts elements at specified positions
# list.pop(1) -> removes the element and displays it 
# list.remove('element') -> removes specified element
# list.sort() -> sorting the elements
```

```python
mylist = ['yethish','chintu','cat','dog','meow']
for  i in enumerate(mylist):
    print(i[0])
```

```python
# one type of usecase for enumerate() 
mylist = ['hi','hello','how','are','you']
mydict  = {}
for i, v in enumerate(mylist):
    mydict[i] = v
for i in mydict.items():
    print(i)
```

```python
# sort is used only on lists
# sorted is used for any iterable

my_tuple = (12,-2,0,89,100)
my_list = list(my_tuple)
print(sorted(my_tuple))
print(sorted(my_list))
```

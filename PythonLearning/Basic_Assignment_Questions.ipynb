```python
# Check Palindrome Number
def palindrome(num):
   return num == num[::-1]
palindrome(input("Enter a number: ").lower())
```

```python
def palindrome2(num):
  original_num = num
  rev = 0
  while(num > 0):
    digit = num % 10
    num = num // 10
    rev = rev * 10 + digit
  return original_num == rev
palindrome2(int(input("Enter number: ")))
```

```python
#Merge 2 list using the condition
def merge_list(list1,list2):
  result_list = []
  for i in list1:
    if i % 2 != 0:
      result_list.append(i)
  for j in list2:
    if j % 2 == 0:
      result_list.append(j)
  return result_list
merge_list([1,2,3,4,5,137,140],[1,2,3,4,5,100,133])
```

```python
#get each digit from a number in the reverse order
def reverse_print(num):
  result = ""
  while(num > 0):
    last_digit = num % 10
    num = num//10
    result = result + str(last_digit)
  return " ".join(result)
reverse_print(6353298)
```

```python
def reverse_print1(num):
  digits = []
  while (num > 0):
    digits.append(str(num % 10))
    num = num // 10
  return " ".join(digits)
```

```python
# Fibonacci series
def fibonacci(n):
    series = []
    first = 0
    second = 1
    for i in range(n):
        series.append(first)
        first, second = second, first + second #tuple unpacking
    return series
```

```python
def fibonacci1(n):
    series = []
    first = 0
    second = 1
    for i in range(n):
        series.append(first)
        temp = first
        first = second
        second = temp + second
    return series
```

```python
print(fibonacci(12))
print(fibonacci1(12))
```

```python
# Check if a given year is a leap year
def leap_year(year):
    if(year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)):
      return "Leap Year"
    else:
      return "Non Leap year"
leap_year(2100)
```

```python
# Checks whether a given number is prime or not
import math
def check_prime(num):
  isprime = True
  if num < 2:
    return False
  for i in range(2, math.floor(math.sqrt(num)) +1):
    if num % i == 0:
      isprime = False
      break
  return isprime
primes = [i for i in range(1,100) if check_prime(i)]
print(primes[0:len(primes):2])
```

```python
# WAP to reverse a given number
num = int(input('Enter a number to reverse: '))
reverse = 0
while num > 0:
  digit = num % 10
  reverse = reverse*10 + digit
  num = num //10
print(f"Reverse of a number is: {reverse}")
```

**Dictionary Methods/ Functions**

Dictionary consists of  Key-Value Pairs enclosed in flower brackets and colon in between of each pair.

It can allow only unique values inside (key) area

-> Keys can be Immutable

-> Values can be of any python objets

```python
# Creating a Dictionary
user_d = {}
d1 = {"Name":"Yethishwar",
        "id": 97,
      "College": "Methodist college of Engineering",
      "grade":"A+",
      "Mobile": "9000900025"}
print(type(d1))
```

```python
#Accessing values with the help of keys
print(d1["Name"])
print(d1["id"])
print(d1["College"])
print(d1["grade"])
print(d1["Mobile"])
```

```python
print(d1)
```

```python
#keys() function returns only keys present in the dictionary
d1.keys()
```

```python
#values() fucntion returns only values present in the dictionary
d1.values()
```

```python
#items() function returns both key-value pairs present in the dictionary
d1.items()
```

```python
# Major use case of using .items() function which is used to retieve key, value separatly and work with them.
for key, value in d1.items():
  print(key)
  print(value)
  print("______________")
```

```python
# we can also access key in the dictionary using keyword called key calling in the loop
for key in d1:
  print(key)
  print(d1[key])
```

```python
# Removes the item in the dictionary based on the provided key and displays the value
d1.pop("id")
```

```python
d1.items()
```

```python
# Removes the value which is present at the end of the dictionary
d1.popitem()
```

```python
# copies the values from d1 to d2
d2 = d1.copy()
```

```python

d1.setdefault("id")
```

```python
print(d1)
```

```python
""" If we wated to add an element into the dictionary if the element already exist means
It is going to display the existing element, otherwise sets the default value to the key provided """
d1.setdefault("sports","VollyBall")
```

```python
print(d1)
```

```python
""" update function is used to update the values to the existing dictionary by passing the values in terms of
keyword arguments or dictionaries or iterable pairs like list, tuples... if the updated key is already presents means
it is going to overwrite the value associated with that key"""
temp_dict = {'color':"aqua", "Blood Group":"B+"}
d1.update(temp_dict)
```

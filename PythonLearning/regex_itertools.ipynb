```python
import re
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!@#$%*?]', '', value) # remove these characters from the string
        value = value.title()
        result.append(value)
    return result
```

```python
clean_strings(['Hi!!!!!!!!!!','YetHisHWAR####@@@@','     chintu$$$$','     *@@worKING'])
```

```python
def remove_everything_except_letters(strings):
    result = []
    for value in strings:
        value = value.strip()
        value  = re.sub('[^a-zA-Z]','',value)
        value = value.title()
        result.append(value)
    return result
```

```python
remove_everything_except_letters(['Hi 9863492 hello ','Yethishwar      11111', 'CH'])
```

```python
def pattern_matching(strings):
    result = []
    for value in strings:
        value = value.strip()
        value  = re.sub('^hi','',value)
        result.append(value)
    return result

    # [a-z] -> Lowercase Letters
    # [A-Z] -> Uppercase Letters
    # [abc] -> any one of a, b, c
    # [^abc]  -> except abc
    # [a-zA-Z] -> alphabets only
    # [a-zA-Z0-9]  -> alphanumeric
    # [\d] -> matches digits
    # [\D -> matches non-digits
    # [\w]  -> matches word character[a-zA-Z0-9]
    # [\W] -> except word characters
    # [\s]  -> matches word spaces
    # [\S]  -> matches non-word spaces
    # [\d{10}]  -> matches the digits which are consicutively 10 like phone numbers
    # [\d{10,}]  -> digits between 10 or more 
    # [a*]  -> matches a-----> ex: a, aa, aaa, aaaa, aaaaa, aaaaaaaaa
    # [a+]  -> does not match empyty string 
    # [a?]  -> zero or more matching 
    # [^hello]  -> only matches start of string
    # [hello$]  -> only matches end of string
```

```python
pattern_matching(['aaaaaaaaaaaahello   hi  ', 'hi    aaaaaayethishwar'])
```

```python
def regex_match_function(mystring):
    return re.match('hello', mystring)
regex_match_function('yethishwar hello')


# re.match("Hello", "Hello world")   # ✅ MATCH
# re.match("Hello", "Say Hello")     # ❌ NO MATCH
# it behaves like ^ implicitly
```

```python
def regex_findall_function(mystring):   # Extract Everything
    return re.findall('\d+', mystring)
regex_findall_function('yethishwar age 21 , 100  for yethish')


# Used to extract texts like email, phone numbers, pincodes...
```

```python
import itertools # it provides fast and memory efficient iterator tools for working with iterations and combinations
# count()	     -> Generates increasing numbers
# cycle()	     -> Repeats elements endlessly
# combinations() -> Unordered unique selections
# permutations() -> Ordered unique arrangements
# accumulate()	 -> Cumulative running results
# chain()	     -> Joins multiple iterables
# groupby()      -> Groups adjacent elements
# product()	     -> Cartesian product combinations
```

```python
# Generate values endlessly (until you stop them).
from itertools import count   

for i in count(0, 10):
    if i > 100:
        break
    print(i)
```

```python
# repeats elements in an iterable forever
from itertools import cycle
for i, ch in enumerate(cycle(['Yethishwar'])):
    if i == 6:
        break
    print(ch)
```

```python
# cumulative operations sum() by default 

from itertools import accumulate
list(accumulate([1, 2, 3, 4]))
```

```python
# COmbines multiple iterables
from itertools import chain
list(chain([1,2],[3,4],[5,6],[3,[1,2]]))
```

```python
list(chain((1,2),(3,4)))
```

```python
list(chain("Yethish",'chintu'))
```

```python
# it groups consecutive elements
from itertools import groupby
data = "aabaac"
result = {}
for key, group in groupby(data):
    print(key, list(group))
```

```python
# cartesian product
from itertools import product
list(product([1,2,3],['A','b','C','D']))
```

```python
# permutations -> order matters
from itertools import permutations
list(permutations([1,2,3],2))
```

```python
# combinations -> order doesnot matters 
from itertools import combinations 
list(combinations([1,2,3],2))
```

```python
# combinations with replacement
from itertools import combinations_with_replacement
list(combinations_with_replacement([1,3,1],2))
```

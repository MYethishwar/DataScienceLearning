```python
import matplotlib.pyplot as plt
# In this data the the man anscombe trying to tell that...
# Even though the summary statistics looks itendical... 
# it looks completly different when plotted
df = sns.load_dataset("anscombe")

sns.lmplot(x="x", y="y", col="dataset", hue="dataset",
           data=df, col_wrap=2, ci=None, palette="muted",
           height=4, scatter_kws={"s": 50, "alpha": 0.8})
plt.show()
group_means = df.groupby("dataset").mean()
group_varience = df.groupby("dataset").var()
print(f"Mean: {group_means}")
print(f"Mean: {group_varience}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([2,5,10,20])
ypoints = np.array([60,70,75,100])
plt.xlabel("Hours Studied")
plt.ylabel("Score Achieved")
plt.title("Hours studied to Score achieved")
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
ypoints = np.array([300, 800, 100, 1000, 500, 700])
plt.plot(ypoints,'o-c', ms=15, mec='black', mfc='y' )  # 'marker, line type, color'
            # We can also include -> markersize as ms, markeredgecolor as mec, markerfacecolor as mfc
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
# We can also use ls=':/-/--/-.'
plt.show()
```

```python
# Line color 
# We can use color or shorter one c to set the color of the line
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

# plt.plot(ypoints, color= '#BF1F92')
plt.plot(ypoints, 'ro-', linewidth='2.0') # We can also do this 
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
# Create Labels for a plot
# plt.xlabel("Here the xlabel will come")
# plt.ylabel("Here the ylabel will come")
# plt.title("Used to give a title")
# We can use pos='left/right/center' in title for alignment
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'Serif','color':'blue','size':25}
font2 = {'family':'Serif','color':'darkred','size':20}

plt.title("Sports Watch Data", fontdict = font1, loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(axis='x', linestyle='--', linewidth=1.5,color='g') # we can specify axis for gridlines
plt.plot(x, y)
plt.show()
```

```python
# Display Multiple plots
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 5, 1) # specify subplot to combine multiple plots.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 5, 3)  # this arguments states that we have 1 row, 5 columns and this is the third plot
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#plot1 1:
plt.subplot(2,1,1)  # this states that we have 2 rows 1 column and this is first plot
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2) # This arguments states that we have 2 rows, 1 column and this is second plot
plt.plot(x,y)

plt.show()
```

```python
# We can draw as much plots as we can
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#plot1 1:
plt.subplot(2,3,1)  
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 2:
plt.subplot(2,3,2) 
plt.plot(x,y, 'r')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 3:
plt.subplot(2,3,3) 
plt.plot(x,y, 'g')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 4:
plt.subplot(2,3,4) 
plt.plot(x,y, 'b')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 5:
plt.subplot(2,3,5) 
plt.plot(x, 'y')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 6:
plt.subplot(2,3,6)   # This means we have 2 rows and three columns 
plt.plot(x,y,'black')
plt.suptitle("This is the suptitle()")
plt.show()
```

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5) # We can mention alpha value for transparency

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')

plt.colorbar()

plt.show()
```

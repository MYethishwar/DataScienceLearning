```python
# Difference betweeen copy() and view()
copied_array = myarray.copy()
print(f"Original array: {myarray}")
print("")
print(f"Copied array: {copied_array}")
copied_array[0] = 1000
print(f"Edited Copied Array: {copied_array}")
print(f"Original array remains same : {myarray}")
print("")
view_array = myarray.view()
print(f"View of array: {view_array}")
view_array[0] = 1000
print(f"Edited view Array: {view_array}")
print(f"Original array : {myarray}")
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr3 = np.array([89,99,100])
```

```python
# cocmbining along columns by default
new_array = np.concatenate((arr1, arr2,arr3))
print(new_array)
```

```python
# specify axis = 1 to combine along rows
new_array = np.concatenate((arr1, arr2,arr3), axis=0)
print(new_array)
```

```python
# Combining using stack function
print(f"Array 1: {arr1}")
print(f"Array 2: {arr2}")
print("Stacked array: ")
stack_array = np.stack((arr1, arr2))
print(stack_array)
concat_array = np.concat((arr1, arr2))
print("Concatenation: ")
print(concat_array)
```

```python
# Applying filter concept based on a condition
raw_array = np.array([1,2,6,56,77,79,74,44,12,0,99,97,57,43,31])
bool_array = []
for element in raw_array:
    if element > 80:
        bool_array.append(True)
    else: 
        bool_array.append(False)
bool_array = np.array(bool_array)
print(bool_array)
print(np.where(bool_array == True))
print(raw_array[bool_array])
```

```python
# Another way of doing
raw_array = np.array([1,2,6,56,77,79,74,44,12,0,99,97,57,43,31])
bool_array = raw_array > 80
print(bool_array)
print(np.where(bool_array == True))
print(raw_array[bool_array])
```

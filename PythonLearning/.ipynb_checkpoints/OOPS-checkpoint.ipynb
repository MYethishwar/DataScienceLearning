```python
class Car:
    def start_engine(self,a):
        print(f"Started Engine Number: {a}")
    def off_engine(self, a):
        print(f"Engined turned off: {a}")
```

```python
myobject = Car()
```

```python
myobject.start_engine("1232")
```

```python
myobject.off_engine(323)
```

```python
class Math:
    def mul(self,a,b):
        return a * b
myob = Math()
myob.mul(12,12)
```

```python
class Student:
    def __init__(self, student_name, marks_scored):
        self.student_name = student_name 
        self.marks_scored = marks_scored
    def get_avg(self):
        return sum(self.marks_scored)/len(self.marks_scored), self.student_name
s = Student("Yethishwar", [10, 88, 90, 99, 87])
```

```python
s.get_avg()
```

```python
class CarDetails:
    def __init__(self,brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def information(self):
        return f"{self.brand} {self.model} {self.year}"
c = CarDetails("BMW", "M5", 2022)
c.information()
```

```python
# instance attributes vs class attributes

class Academy:
    academy_name = "Intellintex" # class attributes

    def __init__(self,name):
        self.name = name
s1 = Academy("yethishwar")
s2 = Academy("Chintu")

print(s1.academy_name)
Academy.academy_name = "Yethish Academy"
s1.academy_name
```

```python
# Instancemethod, staticmethod and classmethod
class Maths:
    pi = 3.14
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return Maths.pi * self.radius ** 2
        
    @classmethod
    def change_pi(cls,new_pi):
        cls.pi = new_pi
        
    @staticmethod
    def add(a,b):
        return a + b


m = Maths(5)
print(m.area())
```

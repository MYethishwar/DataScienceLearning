```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = sm.datasets.get_rdataset("mtcars").data
df.head()
```

```python
df.groupby("cyl")['mpg'].mean().plot(kind='bar',color='green')
plt.title("Cylinder count per miles per galoon")
plt.xlabel("No of cylinders")
plt.ylabel('Miles per gallon')
plt.show()
```

```python
plt.figure(figsize=(10,3))
df['mpg'].plot(kind='line',marker='*',color='green')
```

```python
from matplotlib import colors
gear_counts = df['gear'].value_counts()
gear_counts.plot(kind='pie',autopct='%1.1f%%',colors=['aqua','green','black'])
```

```python
df.plot(kind='scatter',x='wt',y='mpg',marker ='*')
```

```python
df['hp'].plot(kind='hist',color='lightblue',edgecolor='green')
```

```python
plt.figure(figsize=(15,3))
df['mpg'].plot(kind='area',alpha=0.5,color='violet')
```

```python
# Seaborn 
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
sns.histplot(data=df,x='total_bill',color='red')
plt.show()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',size='size')
plt.title('Relationship between TotalBill with TIp and gender and size')
plt.show()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='day',size='size')
plt.grid()
plt.show()
```

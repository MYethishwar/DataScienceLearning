```python
# A Pandas series is like a column in a table
# it is one dimentional array holding data of any type
a = ["Yethishwar","madhavaram","Chintu"]
series_data = pd.Series(a)
print(series_data)
```

```python
# If no labels are specified means the index numbers will be showed in the output by default
series_data[0]
```

```python
# By using index=[] argument we can name our own lalbels
series_data = pd.Series(["yethishwar", 22, "Hyderabad"], index=["name","age","place"])
```

```python
series_data
```

```python
# We can also access the data using our custom labels
print(series_data["name"])
print(series_data["age"])
```

We can also use key value object like a dictionary to when creating Series

The keys of the dictionary will become the labels

```python

calories = pd.Series({'day1':2000, 'day2': 1000, 'day3': 5000})
print(calories)
```

```python
# We can also access specific indiexes to series
claories = {'day1':2000, 'day2': 1000, 'day3': 5000}
data = pd.Series(calories, index=['day2','day3'])
print(data)
```

DATA FRAME


Datasets in pandas usually in multidimentional tables we call them as data frames
It is like a 2 dimentional dataset consists of rows and columns

Series is like a column a dataframe is the whole table

```python
data = {
    "calories": [1000,2000,5000,540,600,200],
    "duration": [20,30,50,112,10,90]
}
dataframe = pd.DataFrame(data)
print(dataframe)
```

```python
# When using [] means it denotes a pandas dataframe
print(dataframe.loc[[0,4,5]])
```

```python
# Names Indixes
my_data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(my_data, index=['day1','day2','day3'])
print(df)
```

```python
# locate named index
df.loc[['day2','day1']]
```

```python
# Cleaning Data
# Bad data could be:
#       Empty cells
#       Data in wrong format
#       Wrong data
#       Duplicates
```

```python
# 1. CLeaning Empty cells
```

```python
# One way deal with empty cells is to remove rows that contain empty cells
df = pd.read_csv('dirtydata.csv')
```

```python
print(df.to_string())
```

```python
new_df = df.dropna()
```

```python
# By default the dropna() will not change the orignal dataset instead it will create another one
# so using inplace = True will change the original dataset
df.dropna(inplace = True)
```

```python
# Another way to deal with empty values is to replace with a new appropiate value
df = pd.read_csv('dirtydata.csv') #Again raw data is imported
# fillna() is used to replace empty cells with these values
```

```python
new_df = df.fillna(1000)  #inplace = true to change the original dataset
```

```python
# used to count the null values in the dataframe
df.isnull().sum()
```

```python
#Replace only for specified columns
df.fillna({'Calories':130})
df.fillna({'Date':'2020/12/22'})
```

```python
# Replacing with the mean() median() and mode()
avg = df['Calories'].mean()
print(avg)
df.fillna({'Calories':avg}, inplace= True)
```

```python
# mean() is used when the data is normally distributed
# median() is used when the data is skewed or have outliers
```

```python
# Cleaning data of wrong format
# it is difficult to analyze the data which is in wrong format
# TO fix this we have 2 options
#         Remove the rows
#         Convert all cells in the columns in the same format


# pandas has a to_datetime() for this
df['Date'] = pd.to_datetime(df['Date'],format='mixed')
```

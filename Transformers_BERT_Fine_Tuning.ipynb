```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification,Trainer,TrainingArguments
```

```python
data = {
    "text":[
        "The course content was excellent and easy to understand",
        "I really enjoyed the teaching style",
        "The Instructor was very boring",
        "The lectures were confusing and poorly structured",
        "Amazing experience, learned a lot",
        "Waste of time, not recommended"
    ],

    "labels":[1,1,0,0,1,0]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Load the tokeinize and model

Model_Name= "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(Model_Name)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    Model_Name,
    num_labels = 2
)
```

```python
# toknize the data

encodings = tokenizer(
    df['text'].tolist(),
    padding=True,
    truncation=True,
    max_length = 128,
    return_tensors = 'pt'
)
```

```python
labels = torch.tensor(df['labels'].tolist())
```

```python
# Create the torch dataset

class CustomDataset(torch.utils.data.Dataset):

  def __init__(self,encodings,labels):
    self.encodings = encodings
    self.labels = labels

  def __getitem__(self,idx):
    item = {key:val[idx] for key, val in self.encodings.items()}
    item['labels'] = self.labels[idx]
    return item

  def __len__(self):
    return len(self.labels)

dataset = CustomDataset(encodings,labels)
```

```python
# Training Arguements

training_args = TrainingArguments(
    output_dir = "./results",
    num_train_epochs = 50,
    per_device_train_batch_size = 2,
    learning_rate = 2e-5,
    logging_steps = 1,
    save_strategy='no',
    #evaluation_strategy='no',
    report_to="none"
)
```

```python
trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = dataset
)
```

```python
# fine tuning the model

trainer.train()
```

```python
model.eval()
```

```python
test_sentences = [
    "good learning enviroment",
    "I did not understand anything"
]
```

```python
test_inputs = tokenizer(
    test_sentences,
    padding=True,
    truncation=True,
    return_tensors='pt'
)
```

```python
import torch
with torch.no_grad():
  outputs = model(**{k: v.to(model.device) for k, v in test_inputs.items()})
```

```python
predictions = torch.argmax(outputs.logits,dim=1)
```
